```python
import numpy as np
```

```python
def func(delta, k):
    return delta ** k * (3 - delta) - 5 * (1 - delta) - delta ** 201
```

```python
count = 0
deltas = np.arange(0.01, 1, 0.01)
ks = np.arange(0, 200, 1)
for d in deltas:
    for k in ks:
        v = func(d, k)
        if v > 0:
            print("Delta: {:.2f} k: {} Result: {:.2f}".format(d, k, v))
            count += 1

print(count)
```
